```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
from scipy.stats import ttest_ind
```

```python
good_runs_crease = np.load('good_runs_crease.npy')
bad_runs_crease = np.load('bad_runs_crease.npy')
good_runs_num_pos = np.load('good_runs_num_pos.npy')
bad_runs_num_pos = np.load('bad_runs_num_pos.npy')
good_runs_sum = np.load('good_runs_sum.npy')
bad_runs_sum = np.load('bad_runs_sum.npy')
good_runs_crease_init = np.load('good_runs_crease_init.npy')
bad_runs_crease_init = np.load('bad_runs_crease_init.npy')
good_runs_num_pos_init = np.load('good_runs_num_pos_init.npy')
bad_runs_num_pos_init = np.load('bad_runs_num_pos_init.npy')
good_runs_sum_init = np.load('good_runs_sum_init.npy')
bad_runs_sum_init = np.load('bad_runs_sum_init.npy')
```

```python
plt.scatter(np.zeros(len(bad_runs_crease_init)), bad_runs_crease_init, color='red', label='bad runs')
plt.scatter(np.ones(len(good_runs_crease_init)), good_runs_crease_init, color='green', label='good runs')
p_val_crease = ttest_ind(bad_runs_crease_init, good_runs_crease_init)[1]
plt.title(f'Crease values: p-value: {p_val_crease}')
plt.xlabel(f'bad mean {np.mean(bad_runs_crease_init)}, good mean {np.mean(good_runs_crease_init)}')
plt.legend()
plt.show()
```

```python
plt.scatter(np.zeros(len(bad_runs_crease)), bad_runs_crease, color='red', label='bad runs')
plt.scatter(np.ones(len(good_runs_crease)), good_runs_crease, color='green', label='good runs')
p_val_crease = ttest_ind(bad_runs_crease, good_runs_crease)[1]
plt.title(f'Crease values: p-value: {p_val_crease}')
plt.xlabel(f'bad mean {np.mean(bad_runs_crease)}, good mean {np.mean(good_runs_crease)}')
plt.legend()
plt.show()
```

```python
plt.scatter(np.zeros(len(bad_runs_num_pos_init)), bad_runs_num_pos_init, color='red', label='bad runs')
plt.scatter(np.ones(len(good_runs_num_pos_init)), good_runs_num_pos_init, color='green', label='good runs')
p_val = ttest_ind(bad_runs_num_pos_init, good_runs_num_pos_init)[1]
plt.title(f'Percent of elements of n > 0: p-value: {p_val}')
plt.xlabel(f'bad mean {np.mean(bad_runs_num_pos_init)}, good mean {np.mean(good_runs_num_pos_init)}')
plt.legend()
plt.show()
```

```python
plt.scatter(np.zeros(len(bad_runs_num_pos)), bad_runs_num_pos, color='red', label='bad runs')
plt.scatter(np.ones(len(good_runs_num_pos)), good_runs_num_pos, color='green', label='good runs')
p_val = ttest_ind(bad_runs_num_pos, good_runs_num_pos)[1]
plt.title(f'Percent of elements of n > 0: p-value: {p_val}')
plt.xlabel(f'bad mean {np.mean(bad_runs_num_pos)}, good mean {np.mean(good_runs_num_pos)}')
plt.legend()
plt.show()
```

```python
plt.scatter(np.zeros(len(bad_runs_sum_init)), bad_runs_sum_init, color='red', label='bad runs')
plt.scatter(np.ones(len(good_runs_sum_init)), good_runs_sum_init, color='green', label='good runs')
p_val = ttest_ind(bad_runs_sum_init, good_runs_sum_init)[1]
plt.title(f'Sum of n: p-value: {p_val}')
plt.xlabel(f'bad mean {np.mean(bad_runs_sum_init)}, good mean {np.mean(good_runs_sum_init)}')
plt.legend()
plt.show()
```

```python
plt.scatter(np.zeros(len(bad_runs_sum)), bad_runs_sum, color='red', label='bad runs')
plt.scatter(np.ones(len(good_runs_sum)), good_runs_sum, color='green', label='good runs')
p_val = ttest_ind(bad_runs_sum, good_runs_sum)[1]
plt.title(f'Sum of n: p-value: {p_val}')
plt.xlabel(f'bad mean {np.mean(bad_runs_sum)}, good mean {np.mean(good_runs_sum)}')
plt.legend()
plt.show()
```

```python
plt.scatter(bad_runs_crease_init, bad_runs_crease, color='red', label='bad runs')
plt.scatter(good_runs_crease_init, good_runs_crease, color='green', label='good runs')
plt.title('Crease values')
plt.xlabel('initial crease')
plt.ylabel('final crease')
plt.legend()
plt.show()
```

```python
plt.scatter(bad_runs_num_pos_init, bad_runs_num_pos, color='red', label='bad runs')
plt.scatter(good_runs_num_pos_init, good_runs_num_pos, color='green', label='good runs')
plt.title('Percent of elements of n > 0')
plt.xlabel('initial percent')
plt.ylabel('final percent')
plt.legend()
plt.show()
```
