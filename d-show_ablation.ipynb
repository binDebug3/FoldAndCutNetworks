```python
def print_dict_structure(d, indent=0):
    for key, value in d.items():
        print("  " * indent + f"{key}")
        if isinstance(value, dict):
            print_dict_structure(value, indent + 1)
        elif isinstance(value, list):
            print("  " * (indent + 1) + "[")
            for item in value:
                if isinstance(item, dict):
                    print_dict_structure(item, indent + 4)
                else:
                    print("  " * (indent + 4) + "-")
            print("  " * (indent + 1) + "]")

# print_dict_structure(train_benchmarking)
# list(set([model.split("_")[0] for model in list(train_benchmarking.keys())]))
# train_benchmarking["3Linear_Cover_1512"]["mean"]["acc"].shape
```
